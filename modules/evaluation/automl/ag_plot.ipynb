```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the base directory
base_folder = Path("2025-06-13/Multiclass")
result_files = list(base_folder.rglob("*results.json"))

f1_data = []
best_f1_data = []

# Extract f1_weighted scores
for file_path in result_files:
    with open(file_path, "r") as f:
        data = json.load(f)

    dataset_name = str(file_path.stem).replace('_Multiclass_results', '')

    # From leaderboard: get best score_val
    lb = data.get("leaderboard", [])
    lb_sorted = sorted(lb, key=lambda x: x.get("score_val", 0), reverse=True)
    if lb_sorted:
        best_f1 = lb_sorted[0].get("score_val")
        best_f1_data.append({"Dataset": dataset_name, "f1_weighted": best_f1})

    # From metrics: get all f1_weighted values
    for entry in data.get("metrics", []):
        f1 = entry.get("f1_weighted")
        if f1 is not None:
            f1_data.append({"Dataset": dataset_name, "f1_weighted": f1})

# Convert to DataFrames
df_all = pd.DataFrame(f1_data)
df_best = pd.DataFrame(best_f1_data)
```

```python
# --- Plot 1: Bar plot of best f1_weighted ---
plt.figure(figsize=(10, 8))
sns.barplot(data=df_best, x="Dataset", y="f1_weighted")
plt.xticks(rotation=60, ha='right')
plt.title("Best f1_weighted (score_val) per Dataset")
plt.ylabel("f1_weighted (score_val)")
plt.tight_layout()
plt.show()
```

```python
# --- Plot 2: Box plot of all f1_weighted scores ---
plt.figure(figsize=(10, 8))
sns.boxplot(data=df_all, x="Dataset", y="f1_weighted")
plt.xticks(rotation=60, ha='right')
plt.title("Distribution of f1_weighted Scores (from metrics) per Dataset")
plt.ylabel("f1_weighted")
plt.xlabel("Dataset")
plt.tight_layout()
plt.show()
```

```python
# --- Plot 3: Violin plot of all f1_weighted scores ---
plt.figure(figsize=(10, 8))
sns.violinplot(data=df_all, x="Dataset", y="f1_weighted", inner="box")
plt.xticks(rotation=60, ha='right')
plt.title("Distribution of f1_weighted Scores per Dataset")
plt.tight_layout()
plt.show()
```

```python
# Combo: Box + Swarm Plot (for raw data points)
plt.figure(figsize=(10, 8))
sns.boxplot(data=df_all, x="Dataset", y="f1_weighted")
sns.stripplot(data=df_all, x="Dataset", y="f1_weighted", color="black", size=3, jitter=True)
plt.xticks(rotation=60, ha='right')
plt.title("Box Plot with Individual f1_weighted Scores (from metrics)")
plt.tight_layout()
plt.show()
```
